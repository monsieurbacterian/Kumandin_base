```python
import re
import seaborn as sns
import numpy as np

from scipy import spatial
from matplotlib import pyplot as plt
```

```python
import pandas as pd
```

```python
blesses = pd.read_excel('Blessings.xlsx')
```

```python
name = blesses['Name'].tolist()
kum = blesses['Kum'].tolist()
rus = blesses['Rus'].tolist()
```

```python
newline = '\n'
for i in range(len(name)):
  print(f'name: {name[i]},{newline} {newline} kum: {kum[i]},{newline} rus: {rus[i]}{newline}')
```

```python
class corpora():
  def __init__(self, name: list, kum: list, rus: list):
    self.text_kum = kum
    self.text_ru = rus
    self.name = name

  def splitting(self, name, bless_kum, bless_rus):
    sent_kum = bless_kum.split('\n')
    sent_rus = bless_rus.split('\n')
    if len(sent_kum) > len(sent_rus):
      razn = len(sent_kum) - len(sent_rus)
      for i in range(razn):
        sent_rus.append('No data')
    elif len(sent_kum) < len(sent_rus):
      diff = len(sent_rus) - len(sent_kum)
      for n in range(diff):
        sent_kum.append('No data')
    print(f'Без пропусков kum: {len(sent_kum)} ru: {len(sent_rus)}')

    return name, sent_kum, sent_rus, bless_kum, bless_rus


  def create_corpora(self):
    data_list = []
    for tail in range(len(self.name)):
      splitted = self.splitting(self.name[tail], self.text_kum[tail], self.text_ru[tail])
      # print(splitted)
      for i in range(len(splitted[1])):
        row = {'Name': splitted[0], 'Sent_kum': splitted[1][i], 'Sent_rus': splitted[2][i], 'Text_kum': splitted[3], 'Text_rus': splitted[4]}
        data_list.append(row)

    df = pd.DataFrame(data_list)

    return df
```

```python
corp = corpora(name, kum, rus)
```

```python
a = corp.create_corpora()
```
